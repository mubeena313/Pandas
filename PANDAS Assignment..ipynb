```python
#1. Create a DataFrame in 5 different ways
#1) FROM A NumPy ARRAY:
import pandas as pd
import numpy as np

data =np.array([['Neha',20,'USA'],
                ['Alan',25,'London'],
                ['Laila', 26, 'korea'],
                ['Jennie',30,'Thailand']])
df=pd.DataFrame(data, columns=['Name', 'Age', 'City'])
df
```

```python
#2) from a dictionary of lists or arrays:
data ={
    'Name': ['Neha','Alan','Laila','Jennie'],
    'Age':[20,25,26,30],
    'City':['USA','London','Korea','Thailand']
}
df =pd.DataFrame(data)
df
```

```python
#3)from a list of tuples:
data =[('Neha',20,'USA'),
       ('Alan',25,'London'),
       ('Laila',26,'Korea'),
       ('Jennie',30,'Thailand')]
df =pd.DataFrame(data, columns=['Name','Age','City'])
df
       
```

```python
# Learn grouping datasets:
#Grouping datasets in pandas involves splitting the data into groups based on some criteria,
#applying a function to each group independently, and then combining the results into a new data structure.\
#This is often useful for performing aggregate operations on subsets of data, especially when dealing with large datasets.
```

```python
#step by step grouping data set:
#1. Creating a DataFrame
import pandas as pd
import numpy as np
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'A', 'B', 'A'],
    'Value': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
}

df = pd.DataFrame(data)
df
```

```python
#2. Grouping by a Column
grouped = df.groupby('Category')
grouped
```

```python
#3. Iterating Over Groups:
for name, group in grouped:
    print("Category:", name)
    print(group)
    print()
```

```python
#4. Applying Aggregate Functions:
sums = grouped.sum()
sums
```

```python
#5. Accessing Specific Groups
group_A = grouped.get_group('A')
group_A
```

```python
#6. Custom Aggregation Functions:
def custom_agg(arr):
    return arr.max() - arr.min()

custom_aggregation = grouped.agg(custom_agg)
custom_aggregation
```

```python
#7. Applying Multiple Functions
aggregated = grouped.agg(['sum', 'mean', 'max'])
aggregated
```

```python
#8. Resetting Index
aggregated = aggregated.reset_index()
aggregated
```
